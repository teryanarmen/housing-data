```python
# Takes data from ./data and plots them on the same chart using matplotlib. */
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import random as rd
import os
import numpy as np
```

```python
# gets all the files in ./data
files = os.listdir("./data")
```

```python
# for each file in ./data plots the data on the same chart
figure(figsize=(12, 8), dpi=160)
for file in files:
    df = pd.read_csv("./data/" + file)
    state = file[0:2]
    plt.plot(df["DATE"], df[state + "STHPI"], label = state)
```

```python
# for each file in ./data plots the data on the same chart if the state is a top 10 performing state
# make the colors easily distinguishable
figure(figsize=(12, 8), dpi=160)
colors = ["red", "blue", "green", "yellow", "orange", "purple", "pink", "brown", "gray", "black", "brown"]
for file in files:
    df = pd.read_csv("./data/" + file)
    state = file[0:2]
    if state in ["CA", "HI", "MA", "MD", "NJ", "NY", "OR", "TX", "WA", "WI"]:
        plt.plot(df["DATE"], df[state + "STHPI"], label = state, color = colors.pop())
plt.legend()
```

```python
# for each file in ./data plots the data on the same chart if the state is a top 10 performing state
# make the colors easily distinguishable
# make the line style easily distinguishable
# only label the x axis for every fifth data point, and only label the year
figure(figsize=(12, 8), dpi=160)
colors = ["red", "blue", "green", "yellow", "orange", "purple", "pink", "brown", "gray", "black", "brown"]
styles = ["solid", "dashed", "dashdot", "dotted"]
for file in files:
    df = pd.read_csv("./data/" + file)
    state = file[0:2]
    if state in ["CA", "HI", "MA", "MD", "NJ", "NY", "OR", "TX", "WA", "WI"]:
        plt.plot(df["DATE"], df[state + "STHPI"], label = state, color = colors.pop(), linestyle = styles[rd.randint(0, 3)])
plt.legend()
plt.xticks(np.arange(0, 300, 60))
plt.show()
```
